```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("/content/heart.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
#correlation matrix plot to understand which features are well correlated
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
df.head()
```

```python
sns.countplot(x='HeartDisease', data=df)
```

```python
df['ExerciseAngina'] = df['ExerciseAngina'].replace({'N':0, 'Y':1})
df
```

```python
map = {'ATA': 1, 'NAP': 2, 'ASY': 3, 'TA': 4}
df['ChestPainType'] = df['ChestPainType'].replace(map)
df
```

```python
map = {'Normal': 1, 'ST': 2, 'LVH': 3}
df['RestingECG'] = df['RestingECG'].replace(map)
df
```

```python
map = {'Flat': 0, 'Up': 1, 'Down': 2}
df['ST_Slope'] = df['ST_Slope'].replace(map)
df
```

```python
map = {'M': 0, 'F': 1}
df['Sex'] = df['Sex'].replace(map)
df
```

```python
feature_set = df.drop('HeartDisease', axis=1, inplace=False)
feature_set.head()
```

```python
target = df['HeartDisease']
target.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(feature_set,target,test_size=0.2,random_state=30)
```

```python
y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=35)
```

```python
knn.fit(X_train,y_train)
```

```python
pred= knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
error_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb_model = GaussianNB()
nb_model = nb_model.fit(X_train, y_train)
```

```python
y_pred = nb_model.predict(X_test)
print(y_pred)
print(classification_report(y_test, y_pred))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg_model = LogisticRegression()
lg_model = lg_model.fit(X_train, y_train)
```

```python
y_pred = lg_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

# SVM

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC()
svm_model = svm_model.fit(X_train, y_train)
```

```python
y_pred = svm_model.predict(X_test)
print(classification_report(y_test, y_pred))
```
